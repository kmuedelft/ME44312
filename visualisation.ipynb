```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import glob
from pandas import json_normalize

# Get all JSON files in the directory
file_paths = glob.glob(r'raw_data_rotterdam/*.json')

# Read and concatenate all JSON files
df_list = [pd.read_json(file) for file in file_paths]
df = pd.concat(df_list, ignore_index=True)

# Normalize the nested JSON data
df_normalized = json_normalize(df['data'])

df_normalized  
```

```python
# create list of vessel name
vessel_name = df_normalized['vessel.name'].unique()
vessel_name

# create seperate dataframe for each vessel
vessel_data = {}
for i in vessel_name:
    vessel_data[i] = df_normalized[df_normalized['vessel.name'] == i]

vessel_data['TRUDIE']
```

```python
# add an extra column called day to the dataframe
for i in vessel_name:
    vessel_data[i] = vessel_data[i].copy()  # Ensure it's a copy
    vessel_data[i]['navigation.time'] = pd.to_datetime(vessel_data[i]['navigation.time'])
    vessel_data[i]['day'] = vessel_data[i]['navigation.time'].dt.day

vessel_data['TRUDIE']
```

```python
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors

# Filter data for the vessel 'TRUDIE'
trudie_data = vessel_data['TRUDIE']

# Create a map centered around the average latitude and longitude of the first day
m = folium.Map(location=[trudie_data['navigation.location.lat'].mean(), trudie_data['navigation.location.long'].mean()], zoom_start=12)

# Get unique days
unique_days = trudie_data['day'].unique()

# Generate a color map
colormap = cm.get_cmap('tab20', len(unique_days))
color_norm = colors.Normalize(vmin=0, vmax=len(unique_days) - 1)

# Add routes for each day
for i, day in enumerate(unique_days):
    day_data = trudie_data[trudie_data['day'] == day]
    coordinates = day_data[['navigation.location.lat', 'navigation.location.long']].values.tolist()
    color = colors.rgb2hex(colormap(color_norm(i)))
    folium.PolyLine(coordinates, color=color, weight=2.5, opacity=1).add_to(m)

# Display the map
m
```
